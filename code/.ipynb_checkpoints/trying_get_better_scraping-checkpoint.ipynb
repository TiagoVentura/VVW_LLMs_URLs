```python
import requests
from datetime import datetime
import pandas as pd
from datetime import date 
from datetime import datetime
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
## get csv
# empty df
combined_df = pd.DataFrame()

# get all files
list_files = os.listdir("../news_us")

# Iterate through all files in the folder
for file_name in list_files:
    # Check if the file is a JSON file
    if file_name.endswith('.json'):
        file_path = os.path.join("../news_us/", file_name)
        # Read the JSON file into a DataFrame
        df = pd.read_json(file_path)
        # Append the DataFrame to the combined DataFrame
        combined_df = pd.concat([combined_df, df], ignore_index=True)

# remove errors from the API
combined_df = combined_df[combined_df["url"]!="https://removed.com"]

# remove empty content
combined_df = combined_df[combined_df["content"]!=""]

# reset index
combined_df.reset_index()

# Remove duplicate rows
combined_df = combined_df.drop_duplicates(subset=['url'])

combined_df.shape
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_text_from_url(url):
    try:
        # Fetch the page content
        response = requests.get(url, timeout=20)
        response.raise_for_status()  # Raises an error for bad responses
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extract all text from the webpage, excluding scripts and styles
        for script in soup(["script", "style"]):
            script.decompose()  # Remove script and style elements
        
        # Get text
        text = soup.get_text(separator=" ")
        
        # Clean up the text by removing extra whitespace
        lines = [line.strip() for line in text.splitlines()]
        text = "\n".join(line for line in lines if line)
        
        return text
    
    except requests.exceptions.RequestException as e:
        return f"Error fetching the URL: {e}"
```

```python
list_content_new = []
for a in combined_df["url"]:
    print(a)
    try:
        list_content_new.append(scrape_text_from_url(a))
    except:
        list_content_new.append("")
```

```python
combined_df["content_scrap2"] = list_content_new
```

```python
combined_df.to_csv("../news_us/aggregated_news_api.csv", index=False)
```
